```python
h={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
```

```python
res= req.get("https://www.melon.com/chart/index.htm",headers=h)
```

```python
soup=bs(res.text,"html.parser")
```

```python
name= soup.select_one(".ellipsis.rank01")
```

```python
print(name.text)
```

```python
lst50= soup.select(".lst50")
print(len(lst50))
```

```python
lst100=soup.select(".lst100")
print(len(lst100))
```

```python
lst= lst50 + lst100
print(len(lst))
```

```python
songlist=[]
singerlist=[]
mood=[]
```

```python
for i in lst:
  name= i.select_one(".ellipsis.rank01 a")
  songlist.append(name.text)
  #print(name.text)
  singer= i.select_one(".ellipsis.rank02 > a")
  singerlist.append(singer.text)
  mood.append('neutral')
  #print(singer.text)
```

```python
import pandas as pd
dic={'name':songlist,'artist':singerlist,'mood':mood}
melon=pd.DataFrame(dic)
```

```python
melon.to_csv('neutral2.csv',encoding='utf-8-sig',index=False)
```
