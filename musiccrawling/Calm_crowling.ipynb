```python
def Crawling1():
    global calmplaylist
    
    for i in range(1, 18):
    
        try: 
            # name, artist crawling
            name_xpath = '//*[@id="frm"]/div/table/tbody/tr[' + str(i) + ']/td[5]/div/div/div[1]/span/a'
            name = driver.find_element_by_xpath(name_xpath).text
            artist_xpath = '//*[@id="frm"]/div/table/tbody/tr[' + str(i) + ']/td[5]/div/div/div[2]/a'
            artist = driver.find_element_by_xpath(artist_xpath).text
        
        except: # if there have many artists, get the first artist only
            artist_xpath = '//*[@id="frm"]/div/table/tbody/tr[' + str(i) + ']/td[5]/div/div/div[2]/a[1]'
            artist = driver.find_element_by_xpath(artist_xpath).text
        
        mood = 'calm'
    
        # 리스트에 저장한다.
        calmplaylist.append([name, artist, mood])
```

```python
# 버튼 클릭하는 함수 pagemove 정의
def pagemove():    
    next_btn = driver.find_element_by_xpath('//*[@id="pageObjNavgation"]/div/span/a')
    actions = webdriver.ActionChains(driver)
    actions.move_to_element(next_btn).pause(1).click().perform()
```
